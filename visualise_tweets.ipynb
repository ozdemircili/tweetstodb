```python
# Get Data from SQLAlchemy query
import pandas as pd
from models import sync, Tweets
session = sync()
# rows = session.query().values(Tweets.id, Tweets.timestamp, Tweets.coordinates)
dicts = [row for row in
    session.query().values(Tweets.id, Tweets.timestamp, Tweets.coordinates)]
df = pd.DataFrame(dicts)
df.index = df[1]
ts = df.resample('D', how='count').unstack(level=1).ix[:, 1:]
ts = ts.rename(
    columns={
        1: 'Total Tweets',
        2: 'Geotagged Tweets'})
ts = ts.sort_index(axis=1)
ax = ts.plot(
    kind="barh",
    stacked="true",
    title="Tweets from @brockleycentral's followers",
    # I happen to like teal and orange
    color=["#ff8300", "#009aaf"])
ax.legend(loc=7)
# default formatting shows hours and gets cut off
ax.set_yticklabels(
    [dt.strftime("%d-%m-%Y") for dt in ts.index.to_pydatetime()])
# magic tick text font size setting from http://bit.ly/1549oI1
plt.setp(ax.get_yticklabels(), fontsize=8)
ax.set_ylabel("Date")
ax.set_xlabel("Number of Tweets")
ax.annotate(
    'Total Tweets: %s' % len(dicts),
    xycoords="axes pixels", textcoords="axes pixels",
    xy=(175, 70), xytext=(175, 70))
```

```python
# Get Data from SQLAlchemy query
import pandas as pd
from models import sync, Tweets
session = sync()
q = session.query(Tweets.id, Tweets.timestamp, Tweets.coordinates)
df = pd.DataFrame([row for row in q])
df.index = df[1]
ct = df[1].count()
ts = df.resample('D', how='count').unstack(level=1).ix[:, 1:]
ts = ts.rename(
    columns={
        1: 'Total',
        2: 'Geotagged'})
ts = ts.sort_index(axis=1)
ax = ts.plot(
    kind="bar",
    stacked="true",
    title="Tweets from @brockleycentral's followers",
    # I happen to like teal and orange
    color=["#ff8300", "#009aaf"])
leg = ax.legend(loc='best', title="Daily Tweet Count")
frame = leg.get_frame()
frame.set_alpha(0.4)
# default formatting shows hours and gets cut off
ax.set_xticklabels(
    [dt.strftime("%d-%m-%Y") for dt in ts.index.to_pydatetime()])
# magic tick text font size setting from http://bit.ly/1549oI1
plt.setp(ax.get_yticklabels(), fontsize=8)
ax.set_ylabel("Number of Tweets")
ax.set_xlabel("Date")
ax.annotate(
    'Total Tweets: {0:,}'.format(ct),
    xycoords="axes pixels", textcoords="axes pixels",
    xy=(3, 210), xytext=(3, 210))
plt.tight_layout()
plt.savefig("example.png", dpi=72)
```
